```python
trips = Table.read_table('http://inferentialthinking.com/notebooks/trip.csv')
trips
```

```python
trips.sort('Duration', descending=True)
```

```python
trips.where('Duration', are.below(1800)).hist('Duration')
```

```python
trips.where('Duration', are.below(1800)).hist('Duration', bins=60, unit='seconds')
```

```python
commute = trips.where('Duration', are.below(1800))
commute.show(2)
```

```python
starts = commute.group('Start Station').sort('count', descending=True)
starts
```

```python
commute.pivot('Start Station', 'End Station')
```

```python
duration = trips.select(3, 6, 1)
duration
```

```python
shortest = duration.group(make_array(0, 1), min)
shortest
```

```python
cc = shortest.where(0, are.containing('Civic Center BART')).sort(2)
cc
```
